```python
import numpy as np
import matplotlib.pyplot as plt

c   = 1
dt = 0.05
dx = 0.1

jmax  = 21
nmax = 6

x = np.linspace(0, dx * (jmax-1), jmax)

# 初期の空間分布を与える. 半分を1, 半分を 0
q = np.zeros(jmax)
for j in range(jmax):
    if (j < jmax / 2):
        q[j]=1
    else:
        q[j]=0


plt.figure(figsize=(8,4),dpi=100)

# 初期分布を書く
plt.plot(x,q, marker ='o')

# 時間発展

for n in range(1, nmax + 1):
    qold = q.copy()  # 同一時間 n における 全q_j をqold に「値渡し」でコピーする

    for j in range(1, jmax -1):
#        q[j] = qold[j]   -c*dt/(2*dx) * (qold[j+1] - qold[j-1])    # Forward-Time-Central-Space
        q[j] = qold[j]   -c*dt/dx * (qold[j] - qold[j-1])               # Upwind

    if(n%2 == 0):
        plt.plot(x, q, marker ='x', label = f'n={n}')

print(q.dtype)

#グラフの後処理
plt.grid(color='black')
plt.xlim([0,2.0])
plt.legend()
plt.show()
```
